```
from sklearn.preprocessing import MinMaxScaler

nbimg=100000

cube = np.zeros((nbimg,101,101,2)) 

for i in range(nbimg):
  
  #original images
  
  hdu = fits.open('imageEUC_VIS-' + str(100000+i) + '.fits')
  data = hdu[0].data[:,:]

  scaler = MinMaxScaler(feature_range=(0, 100))
  data = scaler.fit_transform(data)
  
  cube[i,:,:,0] = data
  
  #rotsub images
  
  hdu = fits.open('imageEUC_VIS-' + str(100000+i) +'_rotsub'+ '.fits')
  data = hdu[0].data[:,:]

  scaler = MinMaxScaler(feature_range=(0, 100))
  data = scaler.fit_transform(data)
  cube[i,:,:,1] = data
#  cube[i,:,:,2] = data

  
  if (i%50==0):
    print (i)

hdu_new = fits.PrimaryHDU(cube)
hdu_new.writeto('CUBEE.fits')

print ("Done")
```
